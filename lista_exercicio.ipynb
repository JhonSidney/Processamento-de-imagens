# Butterworth Passa-baixa

```
img = cv2.imread('Processamento-de-imagens/lista/fotos/image4.jpeg',0)

rows, cols = img.shape
crow,ccol = rows//2 , cols//2 #pega ponto central da imagem

dft = cv2.dft(np.float32(img),flags = cv2.DFT_COMPLEX_OUTPUT) #transformada discreta de fourier
dft_shift = np.fft.fftshift(dft) #mudança do ponto DC

magnitude_spectrum = 20*np.log(cv2.magnitude(dft_shift[:,:,0],dft_shift[:,:,1])) #logarítimo para melhor visualização

```

## cod Butterworth Passa-Baixa n = 5 cutoff = 0.05

```
# Criando o filtro 
x_size, y_size = rows,cols
x_arr, y_arr = np.mgrid[0:x_size, 0:y_size]

cell = (x_size//2, y_size//2)
#matriz com as distâncias
dists = np.sqrt((x_arr - cell[0])**2 + (y_arr - cell[1])**2)

#equacao do butterworth
cutoff =0.05
n=2*5
arg = np.divide(dists,cutoff)**n
filtg = np.divide(1,1+arg);

#cria máscara com o filtro 
mask = np.zeros((rows,cols,2))
mask[:,:,0] = filtg
mask[:,:,1] = filtg
 
fshift = dft_shift*mask

# transformada inversa
f_ishift = np.fft.ifftshift(fshift)
img_back = cv2.idft(f_ishift)
img_back = cv2.magnitude(img_back[:,:,0],img_back[:,:,1])

fig, axes = plt.subplots(1, 3, figsize=(16, 8))
ax = axes.ravel()
ax[0].imshow(magnitude_spectrum, cmap='gray')
ax[0].set_title('Magnitude Spectrum')
ax[0].set_axis_off()

ax[1].imshow(img, cmap='pink')
ax[1].set_title('cavalo')
ax[1].set_axis_off()

ax[2].imshow(img_back, cmap='gray')
ax[2].set_title('Imagem de Saída')
ax[2].set_axis_off()
```

# Butterworth Passa-Alta

##Butterworth Passa-Alta n= 2 cutoff = 0.05

```
img = cv2.imread('Processamento-de-imagens/lista/fotos/image4.jpeg',0)

rows, cols = img.shape
crow,ccol = rows//2 , cols//2 #pega ponto central da imagem

dft = cv2.dft(np.float32(img),flags = cv2.DFT_COMPLEX_OUTPUT) #transformada discreta de fourier
dft_shift = np.fft.fftshift(dft) #mudança do ponto DC

magnitude_spectrum = 20*np.log(cv2.magnitude(dft_shift[:,:,0],dft_shift[:,:,1])) #logarítimo para melhor visualização


cavalo = cv2.imread('Processamento-de-imagens/lista/fotos/image4.jpeg', 0)
rows, cols = cavalo.shape


x_size, y_size = rows,cols
x_arr, y_arr = np.mgrid[0:x_size, 0:y_size]

cell = (x_size//2, y_size//2)
#matriz com as distâncias
dists = np.sqrt((x_arr - cell[0])**2 + (y_arr - cell[1])**2)

#N de  butterworth
N = 2
cutoff = 0.05

u = 2*(N)
arg= np.divide(dists, cutoff)**u;
filtg=np.divide(1,(1+arg)); 

#filtg =  1 - filtg


mask = np.zeros((rows,cols,2))
mask[:,:,0] = filtg
mask[:,:,1] = filtg

mask = 1- mask #PASSA ALTA


fshift = dft_shift*mask
# transformada inversa
f_ishift = np.fft.ifftshift(fshift)
img_back = cv2.idft(f_ishift)
img_back = cv2.magnitude(img_back[:,:,0],img_back[:,:,1])

fig, axes = plt.subplots(1, 2, figsize=(16, 8))
ax = axes.ravel()
ax[0].imshow(cavalo, cmap='gray')
ax[0].set_title('Original')
ax[0].set_axis_off()


ax[1].imshow(img_back, cmap='gray')
ax[1].set_title('Butterworth n= 2  d0 = 0.05')
ax[1].set_axis_off()
```

```
carta = cv2.imread('Processamento-de-imagens/lista/fotos/image7.jpeg')
carta_gray = cv2.imread('Processamento-de-imagens/lista/fotos/image7.jpeg', cv2.IMREAD_GRAYSCALE)

# Convert grayscale image to binary

(thresh, im_bw) = cv2.threshold(carta_gray, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

thresh = 127
im_bw = cv2.threshold(carta_gray, thresh, 255, cv2.THRESH_BINARY)[1]

kernel = np.array([[0,0,0],[1,1,0],[0,0,0]],np.uint8)
carta_erosion = cv2.erode(im_bw,kernel,iterations = 1)

fig, axes = plt.subplots(1, 3, figsize=(20, 10))
ax = axes.ravel() 

ax[0].imshow(cv2.cvtColor(carta, cv2.COLOR_BGR2RGB))
ax[0].set_title('Imagem')
ax[0].set_axis_off()

ax[1].imshow(im_bw, cmap = 'gray')
ax[1].set_title('Imagem Binario')
ax[1].set_axis_off()

ax[2].imshow(carta_erosion, cmap = 'gray')
ax[2].set_title('Imagem Erosão')
ax[2].set_axis_off()
```
